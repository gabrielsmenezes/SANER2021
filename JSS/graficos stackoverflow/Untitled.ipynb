```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
```

```python
df1 = pd.read_csv("android.csv", usecols=[20])
```

```python
df2 = pd.read_csv("spring.csv", usecols=[20])
```

```python
df1["Questions"] = 1
```

```python
df2["Questions"] = 1
```

```python
df1 = df1.groupby(by="Category").sum()
```

```python
df2 = df2.groupby(by="Category").sum()
```

```python
df1 = df1.sort_values("Questions")
```

```python
df2 = df2.sort_values("Questions")
```

```python
df1.to_csv("android_questions.csv")
```

```python
df2.to_csv("spring_questions.csv")
```

```python
pie1 = df1.plot(kind="pie", y="Questions", colormap="Blues", autopct="%1.1f%%", fontsize=25, figsize=(7,10), legend=True, labeldistance=None)
pie1.set_title("Android Questions by Category", fontsize=30)
pie1.set_ylabel("")
pie1.legend(prop={"size":20}, bbox_to_anchor=(0.75,0.1))
# pie1.set_bgcolor("lightslategray")
pie1.figure.savefig("android_questions_category_pie.pdf" )
```

```python
pie2 = df2.plot(kind="pie", y="Questions", colormap="tab20c", autopct="%1.1f%%", fontsize=25, figsize=(7,10), legend=True, labeldistance=None)
pie2.set_title("Spring Questions by Category", fontsize=30)
pie2.set_ylabel("")
pie2.legend(prop={"size":20}, bbox_to_anchor=(0.75,0.1))
pie2.figure.savefig("spring_questions_category_pie.pdf")
```
