```python
def removeEspecialCaracters(sample):
    return sample.replace('\n', '')
```

```python
def getSamples(path):
    samples = []
    with open (path) as file:
        for line in file:
            if "path" in line:
                continue
            line = removeEspecialCaracters(line)
            samples.append(line)
    return samples
```
